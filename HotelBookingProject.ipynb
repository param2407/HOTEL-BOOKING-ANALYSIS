```python
cancel_per = df["is_canceled"].value_counts(normalize=True) * 100
cancel_per
```

```python
plt.figure(figsize=(5,4))
plt.title("Reservation Status Count")
plt.bar(["Not cancelled", "Cancelled"], df["is_canceled"].value_counts(), edgecolor="k", width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="hotel",hue="is_canceled",data=df,palette='Blues')
plt.title("Reservation Status is different Hotels",size=20)
plt.xlabel('Hotels')
plt.ylabel('Number of Reservation')
plt.show()
```

```python
resort_hotel = df[df["hotel"] == "Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True) * 100
```

```python
city_hotel = df[df["hotel"] == "City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True) * 100 
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in city and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel["adr"],label= "Resort HOtel")
plt.plot(city_hotel.index,city_hotel["adr"],label= "City HOtel")
plt.legend(fontsize=20)
plt.show()
```

```python
df["month"] = df["reservation_status_date"].dt.month
plt.figure(figsize=(16,8))
sns.countplot(x="month",hue="is_canceled",data=df)
plt.xlabel("Month" ,fontsize=20)                             
plt.ylabel("Reservation_status_date",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR month")
sns.barplot(x="month",y="adr",data=df[df['is_canceled'] == 1].groupby('month').sum()[['adr']].reset_index())
plt.show()
```

```python
cancelled_data = df[df["is_canceled"] == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("Top 10 countries with Cancelled  Reservation")
plt.pie(top_10_countries,autopct="%.2f", labels=top_10_countries.index)
plt.show()
```

```python
df["market_segment"].value_counts()
```

```python
cancelled_data["market_segment"].value_counts(normalize=True)*100
```

```python
cancelled_df_adr = cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)
```

```python
non_cancelled_data = df[df['is_canceled']==0]
non_cancelled_df_adr = non_cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
non_cancelled_df_adr.reset_index(inplace=True)
non_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)
```

```python
plt.figure(figsize=(20,6))
plt.title("Average Daily Rate",fontsize=30)
plt.ylabel("ADR")
plt.plot(non_cancelled_df_adr['reservation_status_date'],non_cancelled_df_adr["adr"],label="Non cancelled")
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr["adr"],label="cancelled")
plt.legend()
plt.show()
```
